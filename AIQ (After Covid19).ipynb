```python
#importing library

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
#pd.set_option("Display.max_rows", 200000)
```

```python
#collecting datasets

df = pd.read_csv(r"C:/Users/user/OneDrive/Desktop/TheShaadiTimes/AIQ2020.csv")
df
```

```python
#only INDIA
df = df.iloc[329775:363051 , :]
df
```

```python
sns.heatmap(df.corr() , annot = True)
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
i = df[df.Specie == 'wd'].index
j = df[df.Specie == 'o3'].index
k = df[df.Specie == 'co'].index
l = df[df.Specie == 'pm10'].index
m = df[df.Specie == 'dew'].index
n = df[df.Specie == 'humidity'].index
o = df[df.Specie == 'pressure'].index
p = df[df.Specie == 'wind speed'].index
q = df[df.Specie == 'wind-speed'].index
r = df[df.Specie == 'wind gust'].index
s = df[df.Specie == 'wind-gust'].index
t = df[df.Specie == 'precipitation'].index
```

```python
a = [i,j,k,l,m,n,o,p,q,r,s,t]
for i in a:
    df = df.drop(i)
    
df["Specie"].value_counts()
```

```python
#pivoting the table
df1 = df.pivot_table(['median',"max","min"], ['Date', 'City',"Country"], 'Specie')
df1
```

```python
#selecting the median column
df2 =  (df1.iloc[:, df1.columns.get_level_values(0)=='median'])
df2
```

```python

df2["no2"] = df2.iloc[:, df2.columns.get_level_values(1)=='no2']
df2["so2"] = df2.iloc[:, df2.columns.get_level_values(1)=='so2']
df2["pm25"] = df2.iloc[:, df2.columns.get_level_values(1)=='pm25']
df2["temperature"] = df2.iloc[:, df2.columns.get_level_values(1)=='temperature']

df2
```

```python
#dropping the multilevel indexing column

df2 = df2.drop(['no2',"pm25","so2","temperature"], axis=1, level=1)
df2
```

```python
#resetting the multi level index

df2.reset_index(inplace = True)
df2
```

```python
#converting into a single column
df2.columns = ['_'.join(col) for col in df2.columns]
df = df2
```

```python
df
```

```python
sns.heatmap(df.corr() , annot = True)
```

```python
sns.countplot(df["City_"])
plt.xticks(rotation = 90)
```

```python
#Average So2 in a City_
fig= plt.subplots(figsize=(20, 12))
state_wise_max_so2 = df[['City_','so2_']].dropna().groupby('City_').median().sort_values(by='so2_')
sns.barplot(x='so2_', y=state_wise_max_so2.index, data=state_wise_max_so2)
plt.title("Average So2 in a City after lockdown",fontsize = 30)
plt.xlabel("So2",fontsize = 30)
plt.ylabel("City",fontsize = 30)
plt.show()
```

```python
#Average no2 in a City_
fig= plt.subplots(figsize=(20, 12))
state_wise_max_so2 = df[['City_','no2_']].dropna().groupby('City_').median().sort_values(by='no2_')
sns.barplot(x='no2_', y=state_wise_max_so2.index, data=state_wise_max_so2)
plt.title("Average No2 in a City after lockdown",fontsize = 30)
plt.xlabel("No2",fontsize = 30)
plt.ylabel("City",fontsize = 30)
plt.show()
```

```python
#Average pm25 in a City_
fig= plt.subplots(figsize=(20, 12))
state_wise_max_so2 = df[['City_','pm25_']].dropna().groupby('City_').median().sort_values(by='pm25_')
sns.barplot(x='pm25_', y=state_wise_max_so2.index, data=state_wise_max_so2)
plt.title("Average pm25 in a City after lockdown",fontsize = 30)
plt.xlabel("Pm25",fontsize = 30)
plt.ylabel("City",fontsize = 30)
plt.show()
```
